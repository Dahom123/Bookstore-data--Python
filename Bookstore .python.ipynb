```python
if 'category_name' in df.columns and 'revenue' in df.columns:
    category_revenue = df.groupby('category_name')['revenue'].sum()
    plt.figure(figsize=(10, 8)) 
    plt.pie(category_revenue, labels=category_revenue.index, autopct='%1.1f%%', startangle=140)
    plt.title('Revenue Distribution by Category')
    plt.axis('equal')
    plt.show()
else:
    print("The dataframe does not contain the required columns 'category_name' and 'revenue'.")
```

```python
if 'order_date' in df.columns and 'revenue' in df.columns:
    df['order_date'] = pd.to_datetime(df['order_date'])
    df.sort_values(by='order_date', inplace=True)
    plt.figure(figsize=(10, 6))
    plt.plot(df['order_date'], df['revenue'], marker='o', linestyle='-')
    plt.title('Revenue Trend Over Time')
    plt.xlabel('Order Date')
    plt.ylabel('Revenue')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()
else:
    print("The dataframe does not contain the required columns 'order_date' and 'revenue'.")
```

```python
# Visualize the distribution using histograms
numerical_cols = ['Total_unit', 'revenue']
numerical_data = df[numerical_cols]
plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(1, 2, i)
    plt.hist(numerical_data[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Visualize the distribution using box plots
plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(1, 2, i)
    sns.boxplot(y=numerical_data[col], color='lightgreen')
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

```python
# Visualize the distribution using density plots
plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(1, 2, i)
    sns.kdeplot(numerical_data[col], color='salmon', shade=True)
    plt.title(f'Density Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
plt.tight_layout()
plt.show()
```
